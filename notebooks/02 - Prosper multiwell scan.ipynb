# Load dataset

```python
from pandas_overview import DataFrameSummary, DataFrameOverview
import pandas as pd
```

```python
df = pd.read_excel("../data/prosper_oilfield_100.xlsx")
df.head()
```

# Create DataFrameSummary

```python
dfs = DataFrameSummary(df)
```

```python
dfs.columns_types
```

```python
dfs.types
```
